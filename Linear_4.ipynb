$$\begin{pmatrix}

\end{matrix}\right|

\end{matrix}
\end{pmatrix}.$$

Составим уравнение и найдем его корни
$$\begin{cases}
x_{1}+x_{2}-x_{3}-2x_{4}=0, \\
~~~~~~~~~x_{2}-x_{3} -5x_{4}=2, \\
~~~~~~~~~~~~~~~~x_{3}-\frac{3}{2}x_{4}=-2.
\end{cases}$$

Корни уравнения
$$\begin{cases}

\end{cases}$$
