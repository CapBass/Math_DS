Напишите код на Python, реализующий построение графиков:
окружности,
эллипса,
гиперболы.


```python
x = []
y1 = []
y2 = []
a = 17
b = 8
for i in range(-4000, 4001):
    x_ = i/1000
    x.append(x_)
    y1.append(np.sqrt(1 - (x_**2/a**2) * b**2))
    y2.append(-np.sqrt(1 - (x_**2/a**2) * b**2))

plt.plot(x, y1, color='red')
plt.plot(x, y2, color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ellipse')
plt.xlim(-4, 4)
plt.ylim(-4, 4);
```

y(x) = k∙cos(x – a) + b

```python
x = np.linspace(-5, 5, 50)
def plot_cos(x, a, b, k):
    plt.plot(k * np.cos(x - a) + b, label=f'{k} * np.cos(x - {a}) + {b}')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    
```

Напишите код, который будет переводить полярные координаты в декартовы.

```python
def dec_to_polar(r, alpha):
    x = r * np.cos(alpha)
    y = r * np.sin(alpha)
    return x, y 
```

Решите систему уравнений:
y = x2 – 1
exp(x) + x∙(1 – y) = 1


Решите систему уравнений и неравенств:
y = x2 – 1
exp(x) + x∙(1 – y) > 1


```python
from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return y - x**2 + 1,  (-1 + np.exp(x) + x) / x - y

x1, y1 = fsolve(equations, (1, 1))
x2, y2 = fsolve(equations, (10, 10))
x3, y3 = fsolve(equations, (-4, 4))
```

```python
x = np.linspace(-5, 5, 100)
plt.plot(x, x**2 - 1, label='y = x**2 - 1')
plt.plot(x, (-1 + np.exp(x) + x) / x, label='exp(x) + x * (1 - y) = 1')
plt.scatter(x1, y1, label='(x1, y1)')
plt.scatter(x2, y2, label='(x2, y2)')
plt.scatter(x3, y3, label='(x3, y3)')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True);
```
